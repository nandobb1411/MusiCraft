# Explicação de uma MidiFile

### MetaMessages e Messages

MetaMessages são mensagens de midi que não estão relacionadas com a música em si, mas sim com informações sobre a música. Por exemplo, o nome do arquivo, o nome do autor, o tempo da música, etc.

Message é uma mensagem de midi que está relacionada com a música em si. Por exemplo, uma nota, um acorde, um evento de controle, etc.

MetaMessages são feitas para declarar informações sobre a música, enquanto Messages são feitas para representar a música em si.

### Tipos de MetaMessages

track_name é uma mensagem de midi que indica o nome da track.

set_tempo é uma mensagem de midi que indica o tempo da música.

instrument_name é uma mensagem de midi que indica o nome do instrumento.

time_signature é uma mensagem de midi que indica a assinatura de tempo da música.

key_signature é uma mensagem de midi que indica a assinatura de clave da música.

smpty_offsets é uma mensagem de midi que indica o offset da música.

### Tipos de Messages

note_on é uma mensagem de midi que indica que uma nota foi tocada.

note_off é uma mensagem de midi que indica que uma nota foi parada.

control_change é uma mensagem de midi que indica que um evento de controle foi acionado. (ex: mudança de volume, mudança de instrumento, etc.)

```python
messages = []
metamessages = []
for i in range(len(midi_file.tracks[0])):
    if (midi_file.tracks[0][i].type == 'note_on') or (midi_file.tracks[0][i].type == 'note_off') or (midi_file.tracks[0][i].type == 'control_change'):
        messages.append(midi_file.tracks[0][i])
    else:
        metamessages.append(midi_file.tracks[0][i])
```
